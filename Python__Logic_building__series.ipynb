18.COUNT THE DIGITS IN A NUMBER

```python
n=int(input("enter n"))
count=0
while n>0:
  count+=1
  n=n//10
print("count of digits in a number is:",count)
```

```python
def count_digits(n):
  n=abs(n)
  if n==0:
    return 1
  count=0
  while n>0:
    count+=1
    n=n//10
  return count
count_digits(1234567890)
```

19.CHECK WHETHER THE NUMBER IS PALINDROME OR NOT

```python
n=int(input())
original=n
rev=0
while n>0:
  digit=n%10
  rev=rev*10+digit
  n=n//10
if original==rev:
  print("n is palidrome")
else:
  print("n is not palindrome")
```

28 COUNT THE FREQUENCY OF CHARACTERS IN A STRING

SELECTION SORT ALGORITHM

```python

Lst=[56,3,2,78,6,0]
for i in range(len(Lst)):
  min_val=min(Lst[i:])
  min_ind=Lst.index(min_val)
  Lst[i],Lst[min_ind]=Lst[min_ind],Lst[i]
print(Lst)
```
